```python
#N Reinas - Vuelta Atrás()
################################################################################

#Verifica que en la solución parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solución al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(5,solucion=[],etapa=0)
```

# Solucion Divide y venceras





```python
def distancia_minima_recursiva(puntos):
    if len(puntos) <= 3:
        return distancia_minima_fuerza_bruta(puntos)

    mitad = len(puntos) // 2
    punto_medio = puntos[mitad]

    distancia_izquierda, pares_izquierda = distancia_minima_recursiva(puntos[:mitad])
    distancia_derecha, pares_derecha = distancia_minima_recursiva(puntos[mitad:])

    if distancia_izquierda < distancia_derecha:
        distancia_minima = distancia_izquierda
        pares_minimos = pares_izquierda
    else:
        distancia_minima = distancia_derecha
        pares_minimos = pares_derecha

    puntos_cercanos = []
    for punto in puntos:
        if abs(punto[0] - punto_medio[0]) < distancia_minima:
            puntos_cercanos.append(punto)

    puntos_cercanos.sort(key=lambda punto: distancia_entre_puntos(punto,0))

    for i in range(len(puntos_cercanos)):
        for j in range(i + 1, len(puntos_cercanos)):
            distancia_actual = distancia_entre_puntos(puntos_cercanos[i], puntos_cercanos[j])
            if distancia_actual >= distancia_minima:
                break
            if distancia_actual < distancia_minima:
                distancia_minima = distancia_actual
                pares_minimos = (puntos_cercanos[i], puntos_cercanos[j])

    return distancia_minima, pares_minimos

min_diff, pares = distancia_minima_recursiva(random_list)
print(f"La distancia mínima es: {min_diff}")
print(f"Los puntos que producen la distancia mínima son: {pares}")

```

La complejidad de este algoritmo es O(N log N)

ambas soluciones son aplicables a R^n
